```python
#adding dependedncies
import pandas as pd
import numpy as np
```

```python
#importing data
crime_data_df = pd.read_excel('resources/hate_crimes_from_FBI.xlsx')
crime_data_df.head()
```

```python
#removing unneeded columns
crime_data_df = crime_data_df[['incident_id', 'data_year', 'pug_agency_name', 'agency_type_name', 'state_abbr', 
                               'state_name', 'division_name', 'region_name', 'offender_race', 
                               'offender_ethnicity', 'victim_count', 'offense_name', 'total_individual_victims',
                              'location_name', 'bias_desc', 'victim_types', 'multiple_offense', 'multiple_bias']]
crime_data_df.head()
```

```python
#indexing df by year
crime_data_df = crime_data_df.set_index('data_year')
crime_data_df.head()
```

```python
#creating csv file for ethnicity data
ethnicity_types_df = pd.DataFrame(crime_data_df['offender_ethnicity'].unique(), columns=['ethnicity_type'])
ethnicity_types_df.index.name = 'ethnicity_type_id'

print(f"Max column size - ethnicity_type: {ethnicity_types_df['ethnicity_type'].map(lambda x: len(x)).max()}")

ethnicity_types_df
```

```python
#exporting ethnicity dataframe as a csv for upload
ethnicity_types_df.to_csv('resources/ethnicity_types.csv')

# Confirm that export completed
print('Dataframe exported to csv')
```

```python
#creating csv for offenses data
#splitting multiple offenses
filtered_df = crime_data_df[crime_data_df["multiple_offense"]=="M"]
filtered_df["offense_name"] =filtered_df["offense_name"].str.split(';')
explode_df = filtered_df.explode("offense_name", ignore_index=True)
explode_df.head(20)
```

```python
#creating dataframe of offense types
offense_types_df = pd.DataFrame(explode_df['offense_name'].unique(), columns=['offense_type'])
offense_types_df.index.name = 'offense_type_id'

print(f"Max column size - offense_type: {offense_types_df['offense_type'].map(lambda x: len(x)).max()}")

offense_types_df
```

```python
#exporting offenses dataframe as a csv for upload
offense_types_df.to_csv('resources/offense_types.csv')

# Confirm that export completed
print('Dataframe exported to csv')
```

```python
#creating dataframe for incident id numbers
incident_id_df = pd.DataFrame(crime_data_df['incident_id'].unique(), columns=['incident_id'])
incident_id_df.index.name = 'incident_id'

print(f"Max column size - incident_id: {incident_id_df['incident_id'].max()}")

incident_id_df
```
