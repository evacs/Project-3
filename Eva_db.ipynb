```python
import pandas as pd
```

```python
raw_data = pd.read_csv("resources/fbi_hate_crime_data.csv")
raw_data.head()
```

```python
year_df = raw_data[raw_data['data_year'] > 2008].reset_index(drop=True)
year_df.keys()
```

```python
columns_to_drop = ['ori', 'pug_agency_name','pub_agency_unit',
       'agency_type_name','division_name',
       'region_name', 
       'incident_date', 
       'location_name']
drop1_df = year_df.drop(columns=columns_to_drop)
drop1_df.head()
```

```python
pop_code = drop1_df.population_group_code.unique()
pop_code
```

```python
pop_description = drop1_df.population_group_description.unique()
pop_description
```

```python
multiple_code = drop1_df.multiple_bias.unique()
multiple_code
```

```python
columns_to_drop = ['population_group_code', 'population_group_description', 'adult_victim_count', 'juvenile_victim_count',      
         'adult_offender_count',
       'juvenile_offender_count']
drop2_df = drop1_df.drop(columns=columns_to_drop)
drop2_df.keys()
```

```python
target_value = 'S'
single_bias_df = drop2_df[drop2_df['multiple_bias'].str.contains(target_value, case=False)]
single_bias_df.head()
```

```python
target_value = 'M'
multi_bias_df = drop2_df[drop2_df['multiple_bias'].str.contains(target_value, case=False)]
multi_bias_df.head()
```
